## Data Preprocessing - Encoding Categorical Variables

### Subtask:
Identify categorical columns and apply one-hot encoding to convert them into a numerical format suitable for machine learning models. Drop the original categorical columns and the 'patient_id' column.


```python
df_processed = pd.get_dummies(df, columns=['thal'], drop_first=True)
df_processed = df_processed.drop('patient_id', axis=1)
print(df_processed.head())
```

```python
print(df_processed.info())
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(df_processed.corr(), annot=True, cmap='coolwarm')
```

## Data Preprocessing - Feature Scaling

### Subtask:
Scale numerical features using StandardScaler to ensure they contribute equally to the model.


**Reasoning**:
First, I'll separate the features (X) from the target variable (y) and identify the numerical columns to be scaled from the `df_processed` DataFrame. I will then identify the boolean columns that should not be scaled.



```python
X = df_processed.drop('heart_disease_present', axis=1)
y = df_processed['heart_disease_present']
# Identify numerical columns (excluding boolean columns which result from one-hot encoding)
numerical_cols = X.select_dtypes(include=['int64', 'float64']).columns
boolean_cols = X.select_dtypes(include=['bool']).columns

print("Numerical columns to scale:", numerical_cols.tolist())
print("Boolean columns (not scaled):", boolean_cols.tolist())
```

**Reasoning**:
Now that the numerical and boolean columns have been identified, I will import `StandardScaler`, instantiate it, apply it to the numerical features, and then concatenate the scaled numerical features with the unscaled boolean features to create `X_scaled`.



```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled_numerical = scaler.fit_transform(X[numerical_cols])

X_scaled = pd.DataFrame(X_scaled_numerical, columns=numerical_cols, index=X.index)

# Add the boolean columns back to the scaled DataFrame
X_scaled = pd.concat([X_scaled, X[boolean_cols]], axis=1)
print(X_scaled.head())
```

```python
print(X_scaled.info())
```

**Reasoning**:
To split the data into training and testing sets, I need to import the `train_test_split` function and then apply it to `X_scaled` and `y` with the specified parameters.



```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
print("y_train shape:", y_train.shape)
print("y_test shape:", y_test.shape)
```

## Model Training (Logistic Regression)

### Subtask:
Initialize and train a Logistic Regression model using the training data.


**Reasoning**:
I need to import the LogisticRegression model, instantiate it with the specified parameters, and then train it using the X_train and y_train datasets.



```python
from sklearn.linear_model import LogisticRegression

# Instantiate the Logistic Regression model
model = LogisticRegression(random_state=42, solver='liblinear')

# Train the model using the training data
model.fit(X_train, y_train)

print("Logistic Regression model trained successfully.")
```

## Model Evaluation - Predictions

### Subtask:
Make predictions on the test set (X_test) using the trained Logistic Regression model.


**Reasoning**:
To make predictions on the test set, I will use the trained Logistic Regression model's `predict` method on `X_test` and store the output in `y_pred`.



```python
y_pred = model.predict(X_test)
print("Predictions made successfully.")
```

```python
plt.figure(figsize=(6,4))
sns.histplot(y_pred, bins=20, kde=True, color='teal')
plt.title("Predicted Probability Distribution - Logistic Regression")
plt.xlabel("Predicted Probability of Heart Disease")
plt.ylabel("Frequency")
plt.show()
```

## Model Evaluation - Metrics

### Subtask:
Calculate and display key classification metrics such as accuracy, precision, recall, and F1-score to assess the model's performance.


**Reasoning**:
To calculate the specified classification metrics, I need to import the necessary functions from `sklearn.metrics`, then use them with `y_test` and `y_pred` and print the results.



```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

print(f"Accuracy: {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1-score: {f1:.4f}")
```

## Model Evaluation - Confusion Matrix

### Subtask:
Generate and visualize a confusion matrix to understand the types of correct and incorrect predictions made by the model.


```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report

models = {
    "Logistic Regression": LogisticRegression(),
    "Decision Tree": DecisionTreeClassifier(),
    "Random Forest": RandomForestClassifier(),
}

for name, model in models.items():
    model.fit(X_train, y_train)
    preds = model.predict(X_test)
    print(f"{name}: {accuracy_score(y_test, preds):.3f}")
    print(classification_report(y_test, preds))
    print()
```

```python
from sklearn.metrics import confusion_matrix, roc_auc_score, roc_curve

best_model = RandomForestClassifier().fit(X_train, y_train)
y_pred = best_model.predict(X_test)

cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

```python
# Generate and plot ROC curve
y_pred_proba_rf_tuned = best_model.predict_proba(X_test)[:,1]
fpr_rf_tuned, tpr_rf_tuned, _ = roc_curve(y_test, y_pred_proba_rf_tuned)
auc_score_rf_tuned = roc_auc_score(y_test, y_pred_proba_rf_tuned)

plt.figure(figsize=(6,5))
plt.plot(fpr_rf_tuned, tpr_rf_tuned, color='green', linewidth=2, label=f"ROC Curve (AUC = {auc_score_rf_tuned:.2f})")
plt.plot([0, 1], [0, 1], linestyle='--', color='gray')
plt.title("ROC Curve - Optimized Random Forest")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.legend()
plt.show()
```

```python
from sklearn.metrics import roc_curve, roc_auc_score
import matplotlib.pyplot as plt

# Get predicted probabilities for the positive class
y_pred_proba_lr = model.predict_proba(X_test)[:,1]

# Calculate ROC curve for Logistic Regression using probabilities
fpr, tpr, _ = roc_curve(y_test, y_pred_proba_lr)
auc_score = roc_auc_score(y_test, y_pred_proba_lr)

plt.figure(figsize=(6,5))
plt.plot(fpr, tpr, color='navy', linewidth=2, label=f"ROC Curve (AUC = {auc_score:.2f})")
plt.plot([0, 1], [0, 1], linestyle='--', color='gray')
plt.title("ROC Curve - Logistic Regression")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.legend()
plt.show()
```
